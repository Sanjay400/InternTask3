```python
data.drop(columns = 'car_ID', inplace = True)
```

```python
category_column = data.select_dtypes(include= object).columns.tolist()
```

```python
category_column
```

```python
numerical_col = data.select_dtypes(exclude=object).columns.tolist()
```

```python
numerical_col
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()

for column in category_column:
    data[column] = le.fit_transform(data[column])
```

```python
data.sample(3)
```

```python
drop_cols = ['symboling', 'fueltype', 'doornumber', 'enginelocation', 'fuelsystem', 'stroke', 'compressionratio']

data.drop(columns = drop_cols, inplace = True)
```

```python
data.head()
```

```python
X = data.iloc[:, :-1].values
y = data.iloc[:, -1].values
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=2)
```

```python
X_train
```

```python
y_test
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()

X_train= scaler.fit_transform(X_train)
X_test= scaler.transform(X_test)
```

```python
from sklearn.ensemble import RandomForestRegressor

regressor = RandomForestRegressor()
```

```python
regressor.fit(X_train, y_train)
```

```python
prediction = regressor.predict(X_test)
prediction
```

```python
from sklearn.metrics import mean_absolute_error,r2_score
```

```python
mean_absolute_error(y_test, prediction)
```

```python
r2_score(y_test, prediction)
```
